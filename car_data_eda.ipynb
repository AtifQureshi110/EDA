```python
# a function that read csv file meanwhile make copy of original dataset.
def read_csv_and_copy(file_path, new_name):
    print("Reading CSV file...")
    df = pd.read_csv(file_path,encoding='latin-1')
    print("Making a copy of the DataFrame...")
    df_copy = df.copy()
    globals()[new_name] = df_copy
    print(f"Copied DataFrame assigned to variable {new_name}")
    return df_copy
read_csv_and_copy(r"/content/drive/MyDrive/ML/car_data.csv", "car_df")
```

```python
def get_pair_plot(df_id,hue_id):
    sns.pairplot(data= df_id,hue=hue_id)
    """The pairs plot builds on two basic figures, the histogram and the scatter plot.
    The histogram on the diagonal allows us to see the distribution of a single variable.
    while the scatter plots on the upper and lower triangles show the relationship (or lack thereof) between two variables.
    For example, the left-most plot in the second row shows the scatter plot of Age versus User ID."""
    plt.show()
get_pair_plot(car_df,'Purchased')
# data is non-linear
```

```python
def get_percentage_of_col(df, col_id, mas1 = "Percentage of car purchased is: ", mas2 = "Percentage of car not purchased is: "):
    zero = len(df[df[col_id] == 0])
    one = len(df[df[col_id] == 1])
    print(mas1 , "{:.2f}%".format((zero / (len(df[col_id]))*100)))
    print(mas2 , "{:.2f}%".format((one / (len(df[col_id]))*100)))
get_percentage_of_col(car_df, 'Purchased')
# data is imbalanced
```

```python
# it is graphical representation of Purchased the car.
explode = (0.05,0)
car_df.Purchased.value_counts().plot.pie(figsize=(40, 4), autopct='%.1f%%', explode = explode)
plt.title("car Purchase pie diagram")
plt.show()
```

```python
def get_heatmap(df_id):
    """
    heatmap show the relation b/w variables on each axis.
    the Correlation ranges from -1 to +1.
    Values closer to zero means there is no linear trend between the two variables.
    The close to 1 the correlation is the more positively correlated they are;
    that is, as one increases so does the other and the closer to 1 the stronger this relationship is.
    the square's colors also show the relationship such as lighter colors have (more or some) correlation but,
    darker color (less or no) correlation b/w them
    """
    sns.heatmap(df_id.corr(),annot=True)
get_heatmap(car_df)
```

```python
# List of numeric columns
c_list = ['Age', 'AnnualSalary']

# Create a figure with two subplots
fig, axs = plt.subplots(ncols=len(c_list), figsize=(8, 4))

# Loop through the columns
for i, c in enumerate(c_list):
    # Group by "Purchased" and calculate mean or median based on column
    group_data = car_df.groupby("Purchased")[c].agg(["mean", "median"])

    #display(group_data) # if you run this will see the numberic distribution

    # Choose plot type based on column
    plot_type = "Median" if c == "AnnualSalary" else "Mean"

    # Create bar plot for mean or median
    sns.barplot(x=group_data.index, y=plot_type.lower(), data=group_data, ax=axs[i])

    # Set labels and title
    axs[i].set_xlabel("Purchased")
    axs[i].set_ylabel(f"{plot_type} {c}")
    axs[i].set_title(f"{plot_type} {c} by Purchased")

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

- the highest correlation coefficient is between Age and Purchased, with a value of 0.616036, indicating a moderate positive correlation. This suggests that as Age increases, the likelihood of making a purchase also increases.

- The correlation between AnnualSalary and Purchased is lower at 0.364974, indicating a weaker positive correlation. This suggests that there is a relationship between AnnualSalary and Purchased, but it is not as strong as the relationship between Age and Purchased, means it might have change of 36% that if salary increase person may purchase the car.

```python
def get_crosstab(df, col_id1, col_id2, filter_value=None):
    """The crosstab basically take all unique value of col_id1 and separate the value
    of other col according to col_id1 and tell how times they occur"""
    if filter_value is not None:
        df = df[df[col_id1] == filter_value]
    return pd.crosstab(df[col_id1], df[col_id2])

# Get the crosstab for "Purchased" and "Age" columns with filter_value=0
crosstab = get_crosstab(car_df, 'Purchased', 'Age', filter_value=0)
crosstab
```

```python
# Define the age groups
age_groups = [(18, 20), (21, 30), (31, 40), (41, 50), (51, 60), (61, 70)]
# Create a list of purchased and not purchased groups
purchased_groups = [car_df[(car_df['Age'] >= group[0]) & (car_df['Age'] <= group[1]) & (car_df['Purchased'] == 1)] for group in age_groups]
un_purchased_groups = [car_df[(car_df['Age'] >= group[0]) & (car_df['Age'] <= group[1]) & (car_df['Purchased'] == 0)] for group in age_groups]
# Calculate the number of purchases in each age group
num_purchases = [len(group) for group in purchased_groups]
num_un_purchases = [len(group) for group in un_purchased_groups]
# Create a figure with two subplots
fig, axs = plt.subplots(ncols=2, figsize=(10, 4))
# Create bar plots for purchased and unpurchased cars
for i, data in enumerate([(num_purchases, "Number of cars Purchased by Age Group"), (num_un_purchases, "Number of cars not Purchased by Age Group")]):
    sns.barplot(x=[f"{group[0]}-{group[1]}" for group in age_groups], y=data[0], ax=axs[i])
    axs[i].set_xlabel("Age Group")
    axs[i].set_ylabel(data[1])
    axs[i].set_title(data[1])
# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
